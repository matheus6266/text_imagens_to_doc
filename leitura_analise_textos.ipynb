```python
pytesseract.pytesseract.tesseract_cmd = r'C:\\Tesseract\\tesseract' # local de instalação do tesseract
```

```python
# Realizando a leitura dos textos

quantidade_textos = 2
dir_name_images = "C:\\Users\\mathe\\OneDrive\\Pos Graduacao Data Science\\TCC_data_science\\projeto_tio_otavio\\imagens_filtradas"
format_images = ".jpg"

leitura_texto = [] # declarando um vetor vazio

# Preenchendo uma lista com os textos

for i in range(quantidade_textos):
    
    base_filename_image = 'imagem_'+str(i + 1)
    print("Reading de", i + 1 ,"° image")
    leitura_texto.append(pytesseract.image_to_string(os.path.join(dir_name_images, base_filename_image + format_images), lang = 'por', config = '--psm 12')) # Lendo as imagens
    print(leitura_texto[i])
    
    
```
